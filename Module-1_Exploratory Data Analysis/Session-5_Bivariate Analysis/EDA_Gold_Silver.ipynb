```python
import pandas as pd
price_df = pd.read_csv('/content/EDA_Gold_Silver_prices.csv')
price_df.head()
```

```python
corr = price_df.corr()
print(corr)
```

```python
price_df['Year'] = pd.to_datetime(price_df['Month'], format='%b-%y').dt.year

# Filter the DataFrame for the year 2008
df_2008 = price_df[price_df['Year'] == 2008]

# Calculate the correlation between Gold and Silver prices for 2008
corr_2008 = df_2008['GoldPrice'].corr(df_2008['SilverPrice'])

print(corr_2008.round(2))
```
